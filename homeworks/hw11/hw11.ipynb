## Problem 4: Principal Components Analysis

For our final homework problem of the **semester**, we'll apply the concept of principal components analysis to real-world datasets. In addition to reading Chapter 5.4, you should read [this StackExchange post](https://stats.stackexchange.com/questions/2691/making-sense-of-principal-component-analysis-eigenvectors-eigenvalues/140579#140579) for some intuition before proceeding.

Below, we load in the **iris** dataset, which contains measurements of 150 iris flowers, each of which belongs to one of three species. This dataset was collected in 1936 and is one of the most famous datasets in introductory machine learning.

```python
iris = sns.load_dataset('iris')
iris
```

For each flower, we have four numeric features, all measured in centimeters: `sepal_length`, `sepal_width`, `petal_length`, and `petal_width`, in addition to `species`. 

<center><img src="https://content.codecademy.com/programs/machine-learning/k-means/iris.svg" width=500><br><a href="https://content.codecademy.com/programs/machine-learning/k-means/iris.svg">(source)</a></center>

The numeric features constitute a matrix of shape $150 \times 4$, which is too large to visualize. If we wanted to visualize the data, we could use a pairplot, which shows the relationship between each pair of features.

```python
fig = px.scatter_matrix(
    iris,
    dimensions=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'],
    color='species'
)
fig.update_traces(diagonal_visible=False)
fig.show()
```

The dataset has four numeric features, with varying levels of correlation. As we see below, `petal_length` and `petal_width` are very strongly correlated, with a correlation coefficient of $r = 0.96$.

```python
iris.iloc[:, :-1].corr()
```

The main use-case of PCA is to reduce the number of features in the dataset, **not** by selecting a subset of features, but by creating new features that are linear combinations of the original features. These linear combinations are chosen to capture as much of the variation in the data as possible, while being uncorrelated with each other (i.e. having a correlation coefficient of 0), so that each new feature provides information that the earlier new features did not. PCA follows the mantra that variance is information, and so we want to preserve as much of the information as possible. We develop more of the relevant mathematical principles in Chapter 5.4, but for now, let's apply PCA to the iris dataset and interpret the results.

We'll start by centering the data by subtracting each feature's mean from the data.

```python
X = iris.iloc[:, :-1].to_numpy() # Removes the species column, which is not numeric.
X = X - X.mean(axis=0)
X
```

```python
X.shape
```

Next, we'll apply the SVD to the centered data.

```python
u_iris, s_iris, vt_iris = np.linalg.svd(X)
```

The columns of $V$, i.e. the rows of $V^T$, contain the coefficients for the new features.

```python
vt_iris
```

This is telling us that the first new feature, **Principal Component 1**, is given by

$$\text{PC 1} = 0.3613 \cdot \text{sepal length} -0.0845 \cdot \text{sepal width} + 0.8567 \cdot \text{petal length} + 0.3583 \cdot \text{petal width}$$

And the second new feature, **Principal Component 2**, is given by

$$\text{PC 2} = -0.6566 \cdot \text{sepal length} -0.7302 \cdot \text{sepal width} + 0.1734 \cdot \text{petal length} + 0.0755 \cdot \text{petal width}$$

<div class="alert alert-warning">

Make sure you can identify where the numbers in the two formulas above came from!

</div>

The values of Principal Component $i$ can be found by multiplying the centered matrix $X$ by the $i$-th column of $V$, i.e. the $i$-th row of $V^T$:

```python
pc1 = X @ vt_iris[0]
pc1
```

```python
pc2 = X @ vt_iris[1]
```

What do these principal components **look like**? Let's plot PC 2 (on the $y$-axis) against PC 1 (on the $x$-axis), coloring the points by species.

```python
fig = px.scatter(x=pc1, y=pc2, color=iris['species'])
fig.update_layout(xaxis_title='Principal Component 1', yaxis_title='Principal Component 2')
fig.show()
```

Notice that even though we started with four features, this 2-dimensional representation of the data seems to capture the fact that the different species are somewhat distinct and clustered. The colors of each point **were not used** in computing the SVD; the SVD doesn't know anything about the species. To be clear, clusters aren't always guaranteed to be visible in the principal components plot, but they are here!

Also, notice that the resulting scatter plot shows no correlation between the two principal components, as we alluded to earlier.

Here, we computed 2 principal components because it made visualizing the data in a scatter plot easier. There is nothing else special about the choice of 2 principal components. But in practice, we may want to use more than 2 principal components, especially if our goal is to build some predictive model with the data afterwards and we don't need to visualize the data, or if the 2-dimensional scatter plot is not particularly informative.

A more systematic way to choose the number of principal components is to look at the singular values of (the centered) data matrix $X$. As we've seen earlier, the larger a singular value is, the more important the corresponding singular vectors $\vec u_i$ and $\vec v_i$ are to reconstructing $X$.

Specifically,

$$\text{proportion of variance explained by the } i\text{-th principal component} = \frac{\sigma_i^2}{\sum_{j=1}^r \sigma_j^2}$$

where $\sigma_i$ is the $i$-th singular value of $X$, $r = \text{rank}(X)$, and $\sigma_1 \geq \sigma_2 \geq \cdots \geq \sigma_r$.

A graph of the proportion of variance explained by each principal component is known as a **scree plot**.

```python
i = np.arange(1, len(s_iris) + 1)
y = s_iris ** 2 / np.sum(s_iris ** 2)
fig = px.line(
    x=i, y=y, 
    labels={'x': 'Principal Component', 'y': 'Variance Explained'},
)
fig.add_scatter(x=i, y=y, mode='markers', marker=dict(size=8, color='red'), showlegend=False)
fig.update_layout(
    title='Proportion of Variance Explained by Each Principal Component',
    width=600
)
fig.show()
```

What this shows us is that while the data is 4-dimensional, the first principal component alone captures over 92\% of the variance in the data! That is, we can reduce our 4-dimensional dataset to a 1-dimensional dataset and only lose about 8\% of the information in the data. That 1 dimension comes from taking a linear combination of the original four features, which is given by the first column of $V$, i.e. the first row of $V^T$.

Together, the first two principal components capture $92.4\% + 5.3\% = 97.7\%$ of the variance in the data. So, by using only 2 principal components, we're only losing about 2\% of the information in the original 4-dimensional dataset!

In general, one technique for choosing a number of principal components to keep is to pick a number `k` such that the sum of the first `k` principal components captures at least 90\% or 95\% of the variance in the data. Another technique is to pick a `k` at the **elbow** of the scree plot, where the rate of decrease in the proportion of variance explained by the principal components slows down.
